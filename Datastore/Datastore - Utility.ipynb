```python
from google.cloud import datastore

class DS:
  
  def __init__(self,project_id, namespace, kind):
    self.project_id = project_id
    self.namespace = namespace
    self.kind = kind
    self.client = datastore.Client(project_id,namespace=namespace)
    
  def add_new(self, data):
    check_entity = self.filter(series_title=data.get('series_title',''))
    if check_entity != 'Result not found':
      key = self.client.key(self.kind, int(check_entity.key.id))
    else:
      key = self.client.key(self.kind)
    entity = datastore.Entity(key=key)
    entity.update(data)
    self.client.put(entity)
    return entity
  
  def filter(self, **kwargs):
    query = self.client.query(kind=self.kind)
    
    for k, v in kwargs.items():
      query.add_filter(k,'=',v)
    query_result = query.fetch()
    query_result = list(query_result)
    
    if query_result:
      self.result = query_result
      return self
    else:
      return 'Result not found'
  
  def select(self,*cols):
    if not hasattr(self,'result'):
      self = self.filter()
    results = []
    for row in self.result:
      row = {k:v for k,v in row.items() if k in cols}
      results.append(row)
    self.result=results
    return self
    
  def get(self):
    if hasattr(self,'result'):
      if len(self.result)==1:
        result = self.result[0]
      else:
        result = self.result
      del self.result
      return result
    else:
      result = self.filter().get()
      return result
```
